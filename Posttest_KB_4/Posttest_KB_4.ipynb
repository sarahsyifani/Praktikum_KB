Mengatasi nilai null pada kolom year yang bertipe bilangan bulat (integer)

Metode Modus

```python
dataset2 = df.copy()

print("Jumlah Nilai Null pada kolom year sebelum data cleaning >>", dataset2['year'].isna().sum())

dataset2['year'].fillna(dataset2['year'].mode()[0], inplace=True)

print("\nJumlah Nilai Null pada kolom year setelah data cleaning >>", dataset2['year'].isna().sum())
```

Mengatasi nilai null pada kolom type yang bertipe data string

```python
dataset2 = df.copy()

print("Jumlah Nilai Null pada kolom type sebelum data cleaning >>", dataset2['type'].isna().sum())

dataset2['type'].fillna(dataset2['type'].mode()[0], inplace=True)

print("\nJumlah Nilai Null pada kolom type setelah data cleaning >>", dataset2['type'].isna().sum())
```
