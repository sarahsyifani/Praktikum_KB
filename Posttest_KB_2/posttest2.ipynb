3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama

```python
jumlahrecord = len(df)
jumlahattribute = len(df.columns)
daftarattribute = list(df.columns)
attribute_angka = len( df.select_dtypes(include=['number']).columns)
dictionary = {
    'Jumlah record' : jumlahrecord,
    'Jumlah attribute' : jumlahattribute,
    'Daftar attribute' : daftarattribute,
    'Jumlah attribute dengan tipe angka' : attribute_angka
}
for key, value in dictionary.items():
    print(f'{key} : {value}')
```
